```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/home/tejas/retail-sales-project/raw_data/monthly_sales.csv', header=None, delimiter='  ',names=['month', 'sales'])

print(df.head())
print(df.columns)
# Convert month to datetime
df['month'] = pd.to_datetime(df['month'])

# Set style
sns.set_theme(style="whitegrid")

# Plot line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='sales', data=df, marker='o', color='royalblue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
